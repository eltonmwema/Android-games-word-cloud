## Word Cloud Visualizations

From the general view of the data,we realize that the Category column is customized by having the word "GAME" in every detail.A split is hence created to seperate the word from the main detail to have objective data from the category column,and appropriately use it in our intended word cloud visualization.

```python
#Creating the text variable
text = " ".join(cat.split()[1] for cat in df.category)
```

### a.Category Word Cloud

The syntax below represent an intantiation of the WordCloud function

```python
# Creating word_cloud with text as argument in .generate() method
#cloud = WordCloud(width=200,height=150)
word_cloud = WordCloud(collocations = False, background_color = 'black').generate(text)
```

```python
# Display the generated Word Cloud from the category column
fig = plt.figure(figsize=(16, 8))
plt.imshow(word_cloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

### b. Titles Word Cloud

Below we narrow down to have the specific game titles since the above represents the grouped collections.

```python
# Create a wordcloud of the game titles
df['title'] = df['title'].fillna("").astype('str')
title_corpus = ' '.join(df['title'])
title_wordcloud = WordCloud(stopwords=STOPWORDS, background_color='black', height=2000, width=4000).generate(title_corpus)

# Plot the wordcloud
plt.figure(figsize=(16,8))
plt.imshow(title_wordcloud)
plt.axis('off')
plt.show()
```

### NOTE
It is important to realize the dynamics around the wordcloud visualization.

It is many atimes used in sentiment analysis in the business setup where there is a need to have a quick review of the customer feedbacks and feelings about a certain event related to the business enterprise involved.

The survey could involve both open and closed questions,but the wordcloud method is more applicable to the open questions,where there is a wide range of texts to be analysed.

In a generalized perspective,Words represented in a word cloud vary in size at the rate of their count in the corpora commonly known as corpus.i.e,The word that appears the most is the largest in size,and the size diminishes in the order of the number of counts of each word.

In our current case,we have used word cloud to identify the games that appear most and this information could easily be used by stakeholders in a number of ways,among them,games with highest number of subscribers and how the subscription rate is relatable to the revenue inflow.

```python
#Listing the columns available in the dataset
df.columns
```

```python
# Group the game_ratings DataFrame by movie title and calculate the average rating
avg_ratings = df.groupby('title')['total ratings'].mean().sort_values(ascending=False)
avg_ratings
```

```python
#Creating a dictionary of the top 10 rated titles for visualization using a bargraph to avoid avercrowding
top_10 = dict(list(avg_ratings.items())[:10])
top_10
```

### Bargraph representing top ten rated games

```python
plt.figure(figsize=(12, 8))
bar=plt.bar(top_10.keys(), top_10.values(), 
               color=['#4C72B0','#55A868','#C44E52','#8172B2','#CCB974',
               '#64B5CD','#4F4F4F','#4CB391','#BDB76B','#B15455','#6D904F','#7851B8','#A5CCB8',
               '#FFB347','#B276B2','#9F5F9F','#F9B7FF','#F08080','#FFD1DC','#F08080'])

# Add labels to the bars with the total number of movies for each genre
for i, bar in enumerate(bar):
    plt.text(x=i, y=bar.get_height(), s=bar.get_height(), ha='center', va='bottom')

# Customize the plot
plt.xticks(rotation=60, ha='right')
plt.ylabel('Number of ratings')
plt.title('Top 10 ratings by Title')

plt.show();
```

```python
#Dictionary created to represent the bottom ten rated games
bottom_10 = dict(list(avg_ratings.items())[-10:])
bottom_10
```

### Bargraph representing bottom ten rated games

```python
plt.figure(figsize=(12, 8))
bar=plt.bar(bottom_10.keys(), bottom_10.values(), 
               color=['#4C72B0','#55A868','#C44E52','#8172B2','#CCB974',
               '#64B5CD','#4F4F4F','#4CB391','#BDB76B','#B15455','#6D904F','#7851B8','#A5CCB8',
               '#FFB347','#B276B2','#9F5F9F','#F9B7FF','#F08080','#FFD1DC','#F08080'])

# Add labels to the bars with the total number of movies for each genre
for i, bar in enumerate(bar):
    plt.text(x=i, y=bar.get_height(), s=bar.get_height(), ha='center', va='bottom')

# Customize the plot
plt.xticks(rotation=60, ha='right')
plt.ylabel('Number of ratings')
plt.title('last 10 ratings by Title')

plt.show();
```
